```python
import pandas as pd
data = [['diya',57,54,13,56,84],
        ['mahek',40,69,45,55,68],
       ['het',60,52,77,68,72],
       ['vraj',55,42,25,63,85],
       ['shweta',32,26,52,45,20]]
df = pd.DataFrame(data, columns=['Student Name','Maths','English','Hindi','Science','Gujarti'])
df
```

```python
fig, axes = plt.subplots(2,2,layout="constrained")

axes[0,0].plot(df.index,df.Gujarti,'r--', label='guj')
axes[0,0].set_xlabel('x axis')
axes[0,0].set_ylabel('y axis')
axes[0,0].set_title('chart')
axes[0,0].legend()

axes[0,1].plot(df.index,df.Maths,'g*-' , label='math')
axes[0,1].set_xlabel('x axis')
axes[0,1].set_ylabel('y axis')
axes[0,1].set_title('chart')
axes[0,1].legend()

axes[1,0].plot(df.index,df.English,'b-' , label='eng')
axes[1,0].set_xlabel('x axis')
axes[1,0].set_ylabel('y axis')
axes[1,0].set_title('chart')
axes[1,0].legend()


axes[1,1].plot(df.index,df.Hindi,'y^-' , label='hin')
axes[1,1].set_xlabel('x axis')
axes[1,1].set_ylabel('y axis')
axes[1,1].set_title('chart')
axes[1,1].legend()

fig.savefig("filename.png")
```

```python
np.random.seed(3)
x=4+np.random.normal(0,4,24)
y=4+np.random.normal(0,2,len(x))
sizes=np.random.uniform(45,80,len(x))
colors=np.random.uniform(15,80,len(x))

fig, ax = plt.subplots()
ax.scatter(x,y,s=sizes,c=colors,cmap='viridis')
ax.set(xlim=(0,8),ylim=(0,8))
plt.scatter(x,y,s=sizes,c=colors,cmap='brg')
plt.colorbar()
```
