```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
X, y = load_iris(return_X_y=True)
```

```python
y
```

```python
X
```

```python
sse = []
for k in range(1,11):
    km = KMeans(n_clusters=k, init='k-means++' , max_iter=300, n_init=10, random_state=2)
    km.fit(X)
    sse.append(km.inertia_)
```

```python
g=sns.lineplot(x=range(1,11), y=sse)
g.set(xlabel="Number of cluster(k)", ylabel="Sum Squraed Error", title="Elbow Error")
```

```python
kmeans= KMeans(n_clusters = 3 , init='k-means++', max_iter=300,n_init=10,random_state=2)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
pred = kmeans.fit_predict(X)
```

```python
pred
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer =  load_breast_cancer()
```

```python
cancer.keys()
```

```python
print (cancer['DESCR'])
```

```python
import pandas as pd
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

# PCA VISUALIZATON 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca =PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('first principal component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
import seaborn as sns 
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```

```python
import numpy as np 
x=np.arange(10).reshape(5,2)
y=range(5)

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test=train_test_split(x,y)

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)

model.predict(x_test)
```

```python
print("Input" , x)
```

```python
import pickle

with open ('LRmodel.pkl','wb')as f:
    pickle.dump(model,f)
```

```python
with open('LRmodel.pkl', 'rb')as f:

    saved_model=pickle.load(f)
```

```python
saved_model
```

```python
saved_model.predict([[100,101]])
```
