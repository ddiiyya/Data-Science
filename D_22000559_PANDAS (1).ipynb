```python
import pandas as pd
data = [[57,54,13,56,84],
        [40,69,45,55,68],
       [60,52,77,68,72],
        [55,42,25,63,85],
           [32,26,52,45,20]
]
columns = ["Maths", "English", "Hindi", "Science", "Gujarat"]
index = ["diya", "mahek", "het", "vraj", "shweta"]
df = pd.DataFrame(data, columns=columns, index=index)
df['Percentage'] = df.mean(axis=1)
def get_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    elif percentage >= 50:
        return 'E'
    else:
        return 'F'
df['Grade'] = df['Percentage'].apply(get_grade)
df
```

```python
df = pd.DataFrame(data, columns=columns, index=index)
results = pd.DataFrame({
    'Total': df.sum(),
    'Maximum': df.max(),
    'Minimum': df.min(),
    'Standard Deviation': df.std()
})
results
```

```python
sal['BasePay'] = pd.to_numeric(sal['BasePay'], errors='coerce')
sal['OvertimePay'] = pd.to_numeric(sal['OvertimePay'], errors='coerce')
sal['OtherPay'] = pd.to_numeric(sal['OtherPay'], errors='coerce')
sal['Benefits'] = pd.to_numeric(sal['Benefits'], errors='coerce')
sal.info()
```

```python
average = sal['BasePay'].mean()
average
```

```python
highest = sal['OvertimePay'].max()
highest
```

```python
joseph_driscoll = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle'].iloc[0]
joseph_driscoll
```

```python
joseph_driscoll_S = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits'].iloc[0]
joseph_driscoll_S
```

```python
highest_paidperson = sal.loc[sal['TotalPayBenefits'].idxmax()]['EmployeeName']
highest_paidperson
```

```python
lowest_paidperson = sal.loc[sal['TotalPayBenefits'].idxmin()]
lowest_paidperson['EmployeeName']
lowest_paidperson['TotalPayBenefits']
```

```python
average_peryear = sal.groupby('Year')['BasePay'].mean()
average_peryear
```
